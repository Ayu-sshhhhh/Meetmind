# The Multi-Modal Architecture
To handle three different types of data, i will use a **Late Fusion** or **Hybrid Transformer** approach where each input is processed by a specialized "encoder" before being merged.
 1. Input Encoders Sign Language (Video/Landmarks):
 Use __MediaPipe Holistic__ to extract $(x, y, z)$ coordinates of hands, face, and pose. This reduces a heavy video file into a lightweight sequence of landmarks.
    - Pass these through a *Spatio-Temporal GCN* (Graph Convolutional Network) or a *Bi-LSTM.Voice* (Audio):
 2.  Use a pre-trained model like __OpenAI Whisper__ or __Wav2Vec 2.0.__ I will  either use these to convert audio to text immediately or extract "audio embeddings" to keep the model end-to-end.
 3. Text: Use a standard __BERT__ or __RoBERTa encoder__ to create embeddings for text inputs.
 4. The Unified Decoder (The "Brain"):
 Once you have embeddings from all three sources, feed them into a Transformer Decoder.
    - Cross-Attention: The decoder uses attention mechanisms to decide which input (Sign, Voice, or Text) is currently providing the most relevant information to generate the next word in the text output.
